```python
import os
import requests
```

```python
#Define where to save data and URLs
folder_name = "resources"
files_to_download = {
    "ridership_by_station.csv": "https://data.cityofchicago.org/api/views/5neh-572f/rows.csv?accessType=DOWNLOAD",
    "system_info.csv": "https://data.cityofchicago.org/api/views/8pix-ypme/rows.csv?accessType=DOWNLOAD",
    "zoning_districts_2026.csv": "https://data.cityofchicago.org/api/views/dj47-wfun/rows.csv?accessType=DOWNLOAD"
}
```

```python
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    print(f"Created directory: {folder_name}")
```

```python
#If files aren't already downloaded, download them
for file_name, url in files_to_download.items():
    file_path = os.path.join(folder_name, file_name)

    if os.path.exists(file_path):
        print(f"'{file_name}' already exists. Skipping download.")
    else:
        print(f"Downloading {file_name}...")
        try:
            response = requests.get(url, timeout=30)
            response.raise_for_status() # Check for errors

            with open(file_path, "wb") as f:
                f.write(response.content)
            print(f"Successfully saved {file_name}")

        except requests.exceptions.RequestException as e:
            print(f"Failed to download {file_name}: {e}")

print("\nAll set!")
```
